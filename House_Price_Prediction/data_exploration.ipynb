```python
import plotly.graph_objs as go
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
```

```python
df = pd.read_csv('./dataset/train.csv',index_col='Id')
df
```

```python
fig_box = go.Figure()
fig_box.add_box(x=df['SalePrice'],name='Sale_Price')
fig_box.layout.height=300
fig_box.layout.xaxis.range = [0,800000]
fig_box.layout.xaxis.tickfont.size = 13
fig_box.layout.yaxis.tickfont.size = 13
fig_box.show()

fig_hist = go.Figure()
fig_hist.add_histogram(x=df['SalePrice'])
fig_hist.layout.xaxis.range = [0,800000]
fig_hist.layout.xaxis.title = 'Sale_Price'
fig_hist.layout.yaxis.title = 'Count'
fig_hist.layout.yaxis.tickfont.size = 13
fig_hist.layout.xaxis.tickfont.size = 13
fig_hist.show()
```

```python
dtypes = df.dtypes
na_count = df.isna().sum()
zero_count =  df.apply(axis='index',func=lambda x:(x==0).sum())

dtypes.name = 'Dtype'
na_count.name = 'NA_Count'
zero_count.name = 'Zero_Count'

res = pd.merge(dtypes,na_count,left_index=True,right_index=True).merge(zero_count,left_index=True,right_index=True)
res.index.name = 'Feature'
res.sort_values(by=['NA_Count','Zero_Count'],ascending=[False,False])
```

```python
df_feature_sorted = pd.DataFrame([(col,df[col].nunique()) for col in df.columns]).sort_values(by=1,ascending=False)
df_feature_sorted.set_index(0,inplace=True)
```

```python
# writing 'feature' vs 'SalePrice' Scatter plots in a seperate file. size approx 300MB.
with open('feature_explore_level_0.html','wt') as html_file:
    for col in df_feature_sorted.index:
        fig = go.Figure()
        fig.add_scatter(x=df[col], y=df['SalePrice'], mode='markers')
        fig.layout.template = 'gridon'
        fig.layout.xaxis.title = col
        fig.layout.yaxis.title = 'SalePrice'
        fig.layout.yaxis.rangemode = 'tozero'
        fig.layout.title = f'{col} vs SalePrice Nunique={df_feature_sorted.loc[col,1]}'
        html_file.write(fig.to_html())
```
