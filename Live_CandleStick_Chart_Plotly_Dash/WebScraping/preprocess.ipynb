```python
import os
import pandas as pd

col_validator = ['Total Revenue Qtr cr', 'Operating Expenses Qtr Cr',
                 'Operating Profit Qtr Cr', 'Depreciation Qtr Cr', 'Interest Qtr Cr',
                 'Profit Before Tax Qtr Cr', 'Tax Qtr Cr', 'Net Profit Qtr Cr']

for file in os.listdir(path='./nifty_500_scraped_data_raw'):
    df = pd.read_csv(f'./nifty_500_scraped_data_raw/{file}')

    print(f'processing {file}')

    df_transposed = df.set_index('Indicator').transpose()
    df_transposed = df_transposed[col_validator]

    for col in df_transposed.columns:
        df_transposed[col] = df_transposed[col].astype(float)

    df_transposed.index = pd.to_datetime(df_transposed.index,format='%b-%Y')
    df_transposed.sort_index(inplace=True)
    df_transposed.index = df_transposed.index.strftime('%b-%Y')
    df_transposed.to_csv(f'./nifty_500_scraped_data_processed/{file}')
```
